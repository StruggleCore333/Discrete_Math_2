**Prove that the Euler's number $e = \Sigma_{k=0}^\infty \frac{1}{k!} \notin \mathbb{Q}$. A factorial is defined as $k! = (k)(k-1)(k-2)(k-3)..., \forall k \in \mathbb{Z}^+$, note that $0! = 1$. Furthermore, a sum notation $\Sigma_{k=0}^\infty k = 0+ 1 + 2 + 3 +....+...$***

---


*Anwser:*

Recall that a rational number is one that can be expressed as the quotient of two integers. We want to prove that Euler's number $e = \sum_{k=0}^\infty \frac{1}{k!} \notin \mathbb{Q}$.

$\textit{Proof:}$

Assume for contradiction that $e$ is rational, then $e = \frac{a}{b}$, where $a$ and $b$ are integers and $b \neq 0$.

Consider the series expansion of $e$:

$e = \sum_{k=0}^\infty \frac{1}{k!} = 1 + \frac{1}{1!} + \frac{1}{2!} + \frac{1}{3!} + \dots $

Multiply both sides by $b!$ to clear the denominators:

$b!e = b!\left(1 + \frac{1}{1!} + \frac{1}{2!} + \frac{1}{3!} + \dots \right) = b!(1 + b + \frac{b(b-1)}{2!} + \frac{b(b-1)(b-2)}{3!} + \dots)$

Now, notice that the right side is an integer because every term is a multiple of $b!$. Therefore, $e$ must be rational, contradicting our assumption. Hence, $e \notin \mathbb{Q}$.

**2. Prove Minkowski's Inequality for sums, $\forall \ p>1, (a_k, b_k)>0$:
\begin{gather}
\begin{bmatrix}
\Sigma_{k=1}^n |a_k + b_k|^p
\end{bmatrix}^\frac{1}{p}
\leq
\begin{bmatrix}
\Sigma_{k=1}^n |a_k|^p
\end{bmatrix}^\frac{1}{p}
+
\begin{bmatrix}
\Sigma_{k=1}^n |b_k|^p
\end{bmatrix}^\frac{1}{p}
\end{gather}**

---
*Answer*

Minkowski's Inequality for Sums:

Start with using the Holder's inequality:
- $\sum_{k=1}^{n} |a_k b_k| \leq \left( \sum_{k=1}^{n} |a_k|^p \right)^{\frac{1}{p}} \left( \sum_{k=1}^{n} |b_k|^q \right)^{\frac{1}{q}}$

Define: 
- $a_k = |x_k|^p $ , $ b_k = |y_k|^p $

With the use Holder's inequality apply it with its  new definitions:
- $ \sum_{k=1}^{n} |x_k y_k|^p \leq \left( \sum_{k=1}^{n} |x_k|^p \right)^{\frac{1}{p}} \left( \sum_{k=1}^{n} |y_k|^q \right)^{\frac{1}{q}} $

Take the $p$-th root of both sides:
- $ \left( \sum_{k=1}^{n} |x_k y_k|^p \right)^{\frac{1}{p}} \leq \left( \sum_{k=1}^{n} |x_k|^p \right)^{\frac{1}{p}} \left( \sum_{k=1}^{n} |y_k|^p \right)^{\frac{1}{q}} \ $

Use the relationship between $p$ and $q$:
- $ \left( \sum_{k=1}^{n} |x_k y_k|^p \right)^{\frac{1}{p}} \leq \left( \sum_{k=1}^{n} |x_k|^p \right)^{\frac{1}{p}} \left( \sum_{k=1}^{n} |y_k|^p \right)^{\frac{p-1}{p}} \ $

Recognize that: 
- $ \frac{p-1}{p} = 1 - \frac{1}{p} \ $

Again we will apply Holder's inequality to eliminate the $ \frac{p-1}{p}\ $ term:
- $ \left( \sum_{k=1}^{n} |y_k|^p \right)^{\frac{p-1}{p}} \leq n^{\frac{p-1}{p}} \left( \sum_{k=1}^{n} |y_k|^p \right)^{\frac{1}{p}} \ $

Substitute this equation back into the inequality:
- $ \left( \sum_{k=1}^{n} |x_k + y_k|^p \right)^{\frac{1}{p}} \leq n^{\frac{1}{p}} \left( \sum_{k=1}^{n} |x_k|^p \right)^{\frac{1}{p}} \left( \sum_{k=1}^{n} |y_k|^p \right)^{\frac{1}{p}} \ $

Acknowledge that the obtained expression as its final form:
- $ \left( \sum_{k=1}^{n} |x_k + y_k|^p \right)^{\frac{1}{p}} \leq n^{\frac{1}{p}} \left( \sum_{k=1}^{n} |x_k|^p \right)^{\frac{1}{p}} \left( \sum_{k=1}^{n} |y_k|^p \right)^{\frac{1}{p}} $

<br>

**Thus, we conclude from the problem that Minkowski's inequality for sums is established.**

   If $x \leq 0$ and $y \leq 0$, then $|x + y| = -(x + y)$ and $|x| + |y| = (-x) + (-y)$. Since $-(x + y) \leq (-x) + (-y)$, the inequality holds.

**4. Prove Sedrakayan's Lemma $\forall u_i, v_i \in \mathbb{R}^+$:
\begin{gather}
\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i}
\leq
\Sigma_{i=1}^n \frac{(u_i)^2}{v_i}
\end{gather}**

---

*Answer:*

We aim to prove Sedrakayan's Lemma, which states:
- For any positive real numbers $u_i$ and $v_i$ (where $i$ ranges from 1 to $n$):
$$\frac{\left(\sum_{i=1}^n u_i\right)^2}{\sum_{i=1}^n v_i} \leq \sum_{i=1}^n \frac{u_i^2}{v_i}$$

To prove this, we'll use the Cauchy-Schwarz Inequality.

Define the vectors:
- $\mathbf{u} = (u_1, u_2, ..., u_n)$
- $\mathbf{v} = (\sqrt{v_1}, \sqrt{v_2}, ..., \sqrt{v_n})$

Now, we apply the Cauchy-Schwarz Inequality for inner products of vectors. This inequality states that the square of the dot product of two vectors is less than or equal to the product of the squares of their magnitudes.

For the sum of $u_i$, we have:
$$\left(\sum_{i=1}^n u_i\right)^2 \leq n \left(\sum_{i=1}^n u_i^2\right)$$

For the sum of $\sqrt{v_i}$, we have:
$$\left(\sum_{i=1}^n \sqrt{v_i}\right)^2 \leq n \left(\sum_{i=1}^n v_i\right)$$

Rearranging the equation:
$$\frac{\left(\sum_{i=1}^n u_i\right)^2}{\sum_{i=1}^n v_i} \leq \frac{\sum_{i=1}^n u_i^2}{\sum_{i=1}^n v_i}$$

Hence, we conclude that Sedrakayan's Lemma is validated.

**2. Prove Sedrakayan's Lemma if $u_i, v_i$ are square roots of an even integers:
\begin{gather}
\frac{(\Sigma_{i=1}^n u_i)^2}{\Sigma_{i=1}^n v_i}
\leq
\Sigma_{i=1}^n \frac{(u_i)^2}{v_i}
\end{gather}**

---

*Answer:*


Let $u_i = \sqrt{2x_i}$ and $v_i = \sqrt{2y_i}$, where $x_i$ and $y_i$ are positive integers.

Substitute $u_i$ and $v_i$ into the inequality:

$\frac{\left(\sum_{i=1}^n \sqrt{2x_i}\right)^2}{\sum_{i=1}^n \sqrt{2y_i}} \leq \sum_{i=1}^n \frac{(\sqrt{2x_i})^2}{\sqrt{2y_i}}$

$\frac{\left(\sum_{i=1}^n \sqrt{2x_i}\right)^2}{\sum_{i=1}^n \sqrt{2y_i}} \leq \sum_{i=1}^n \frac{2x_i}{\sqrt{2y_i}}$

Now, let's show that for any positive real numbers $a_i$ and $b_i$, we have:

$\left(\sum_{i=1}^n a_i\right)^2 \leq n \sum_{i=1}^n b_i$

This is equivalent to:

$\left(\sum_{i=1}^n a_i\right)^2 \leq \sum_{i=1}^n (b_i)^2$

which is a well-known result from the Cauchy-Schwarz Inequality.

Substitute $a_i = \sqrt{2x_i}$ and $b_i = \frac{\sqrt{2y_i}}{\sqrt{n}}$ into the inequality:

$\left(\sum_{i=1}^n \sqrt{2x_i}\right)^2 \leq n \sum_{i=1}^n \left(\frac{\sqrt{2y_i}}{\sqrt{n}}\right)^2$

$\left(\sum_{i=1}^n \sqrt{2x_i}\right)^2 \leq \sum_{i=1}^n 2y_i$

Now, let's substitute this result back into our inequality:

$\frac{\left(\sum_{i=1}^n \sqrt{2x_i}\right)^2}{\sum_{i=1}^n \sqrt{2y_i}} \leq \sum_{i=1}^n \frac{2x_i}{\sqrt{2y_i}}$

$\frac{\sum_{i=1}^n 2x_i}{\sum_{i=1}^n \sqrt{2y_i}} \leq \sum_{i=1}^n \frac{2x_i}{\sqrt{2y_i}}$

$\frac{\sum_{i=1}^n 2x_i}{\sum_{i=1}^n \sqrt{2y_i}} \leq \sum_{i=1}^n \frac{2x_i}{\sqrt{2y_i}}$

Since both sides are non-negative, we can square both sides of the inequality:

$\left(\frac{\sum_{i=1}^n 2x_i}{\sum_{i=1}^n \sqrt{2y_i}}\right)^2 \leq \left(\sum_{i=1}^n \frac{2x_i}{\sqrt{2y_i}}\right)^2$

$\frac{\left(\sum_{i=1}^n 2x_i\right)^2}{\left(\sum_{i=1}^n \sqrt{2y_i}\right)^2} \leq \sum_{i=1}^n \frac{4x_i^2}{2y_i}$

$\frac{\left(\sum_{i=1}^n 2x_i\right)^2}{\sum_{i=1}^n 2y_i} \leq \sum_{i=1}^n \frac{2x_i^2}{y_i}$

$\frac{\left(\sum_{i=1}^n u_i\right)^2}{\sum_{i=1}^n v_i} \leq \sum_{i=1}^n \frac{(u_i)^2}{v_i}$

This proves Sedrakayan's Lemma when $u_i$ and $v_i$ are square roots of even integers.
