Consider the following experiment:

1. The experiment consists of $n$ identical trials.
2. The outcome of each trial falls into one of $k$ classes or cells.
3. The probability that the outcome of a single trial falls into cell $i$, is $p_i,i = 1,2,...,k$ and remains the same from trial to trial. Notice that $p_1+p_2+p_3+···+p_k =1.$
4. The trials are independent.
5. The random variables of interest are $Y_1 , Y_2 , . . . , Y_k$, where $Y_i$ equals the number of trials for which the outcome falls into cell $i$. Notice that $Y_1 +Y_2 +Y_3 +···+Y_k =n$.
6. The joint probability distribution $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$

Question 1: Show that the expected value $E[y] = \Sigma y \cdotp P(y) = n \cdot p_i$

---

***Answer:***

$E[Y_i] = \sum y \cdot P(y) = n \cdot p_i$

To show that the expected value $E[Y_i] = n \cdot p_i$, we can follow these steps:

**Step 1: Defining the expected value:**

The expected value of $Y_i$ is given by:

$E[Y_i] = \sum_{y_i} y_i \cdot P(Y_i = y_i)$

**Step 2: Use the Joint Distribution:**

The joint probability distribution of $Y_1, Y_2, \ldots, Y_k$ is:

$P(y_1, y_2, \ldots, y_k) = \frac{n!}{y_1! y_2! \cdots y_k!} p_1^{y_1} p_2^{y_2} \cdots p_k^{y_k}$

where $y_1 + y_2 + \cdots + y_k = n$.

**Step 3: Marginal Distribution of $Y_i$:**

Since the trials are independent and follow the multinomial distribution, the marginal distribution of $Y_i$ is binomial:

$Y_i \sim \text{Binomial}(n, p_i)$

The binomial probability mass function is:

$P(Y_i = y_i) = \binom{n}{y_i} p_i^{y_i} (1 - p_i)^{n - y_i}$

**Step 4: Expected Value of Binomial Random Variable:**

For a binomial random variable $Y_i$ with parameters $n$ and $p_i$, the expected value is well known:

$E[Y_i] = n \cdot p_i$

**Thus, we have:**

$E[Y_i] = n \cdot p_i$

For completeness, let’s derive this using the definition of expected value explicitly.

**Step 5: Derivation Using Definition:**

$E[Y_i] = \sum_{y_i=0}^{n} y_i \cdot P(Y_i = y_i)$

Substitute the binomial probability mass function:

$E[Y_i] = \sum_{y_i=0}^{n} y_i \cdot \binom{n}{y_i} p_i^{y_i} (1 - p_i)^{n - y_i}$

**Step 6: Simplify the Summation:**

$E[Y_i] = \sum_{y_i=0}^{n} y_i \cdot \frac{n!}{y_i!(n - y_i)!} p_i^{y_i} (1 - p_i)^{n - y_i}$

Notice that $y_i \cdot \frac{n!}{y_i!(n - y_i)!} = n \cdot \frac{(n-1)!}{(y_i - 1)! (n - y_i)!} = n \cdot \binom{n-1}{y_i-1}$, so we get:

$E[Y_i] = n \cdot \sum_{y_i=0}^{n} \binom{n-1}{y_i-1} p_i^{y_i} (1 - p_i)^{n - y_i}$

**Step 7: Adjust the Summation Index:**

Let $j = y_i - 1$. Then the summation becomes:

$E[Y_i] = n \cdot \sum_{j=-1}^{n-1} \binom{n-1}{j} p_i^{j+1} (1 - p_i)^{n-1-j}$

Notice that when $j = -1$, the term is zero because the binomial coefficient $\binom{n-1}{-1} = 0$.

Therefore, we can write:

$E[Y_i] = n \cdot p_i \sum_{j=0}^{n-1} \binom{n-1}{j} p_i^{j} (1 - p_i)^{n-1-j}$

**Step 8: Recognize the Binomial Theorem:**

The summation is a binomial expansion:

$\sum_{j=0}^{n-1} \binom{n-1}{j} p_i^j (1 - p_i)^{n-1-j} = (p_i + (1 - p_i))^{n-1} = 1$

**Therefore:**

$E[Y_i] = n \cdot p_i \cdot 1 = n \cdot p_i$

**Thus,** the expected value $E[Y_i]$ is indeed $n \cdot p_i$.

${E[Y_i] = n \cdot p_i}$

**Question 1: From the following steps shown above, derive the probability distribution of the experiment.***

*Answers:*

**Derivation of Binomial Distribution**

We aim to derive the probability distribution of the random variable $Y$.

The random variable $Y$ follows a binomial distribution because:

- There are a fixed number of trials, $n$.
- Each trial has two possible outcomes.
- The probability of success $p$ is constant for each trial.
- The trials are independent.

The probability mass function (PMF) of a binomially distributed random variable $Y$ is given by:

$$
P(Y = k) = \binom{n}{k} p^k (1-p)^{n-k}
$$

for $k = 0, 1, 2, \ldots, n$.

Here, $\binom{n}{k}$ is the binomial coefficient and is calculated as:

$$
\binom{n}{k} = \frac{n!}{k!(n-k)!}
$$

Therefore, the probability distribution of $Y$ is:

$$
P(Y = k) = \frac{n!}{k!(n-k)!} p^k (1-p)^{n-k}
$$

for $k = 0, 1, 2, \ldots, n$.


***Question 2: Show that the expectation of this probability distribution is $E[y] = \Sigma y \cdotp P(y) = np$.***

---

*Answers:*

To show that the expectation $E[Y]$ of a binomial random variable $Y$ with parameters $n$ and $p$ is equal to $np$, we start with the definition of the expectation:

$
E[Y] = \sum_{k=0}^{n} k \cdot P(Y = k)
$

Given that $Y$ follows a binomial distribution, the probability mass function is:

$
P(Y = k) = \binom{n}{k} p^k (1-p)^{n-k}
$

### Thus, 
the expectation can be written as:

$
E[Y] = \sum_{k=0}^{n} k \cdot \binom{n}{k} p^k (1-p)^{n-k}
$

To simplify this, we recognize that:

$
\binom{n}{k} = \frac{n!}{k!(n-k)!}
$

Rewriting the expectation:

$
E[Y] = \sum_{k=0}^{n} k \cdot \frac{n!}{k!(n-k)!} p^k (1-p)^{n-k}
$

We can factor out an $n$ from the binomial coefficient:

$
E[Y] = \sum_{k=0}^{n} k \cdot \frac{n}{k} \cdot \frac{(n-1)!}{(k-1)!(n-k)!} p^k (1-p)^{n-k}
$

Simplifying the terms:

$
E[Y] = n \sum_{k=1}^{n} \frac{(n-1)!}{(k-1)!(n-k)!} p^k (1-p)^{n-k}
$

Next, we make a change of index in the summation. Let $j = k-1$. Then when $k$ ranges from $1$ to $n$, $j$ ranges from $0$ to $n-1$. Substituting $k = j+1$:

$
E[Y] = n \sum_{j=0}^{n-1} \frac{(n-1)!}{j!(n-1-j)!} p^{j+1} (1-p)^{n-1-j}
$

This can be written as:

$
E[Y] = n p \sum_{j=0}^{n-1} \frac{(n-1)!}{j!(n-1-j)!} p^j (1-p)^{n-1-j}
$

Recognize that the summation is now the binomial expansion of $(p + (1-p))^{n-1}$, which equals $1$:

$
\sum_{j=0}^{n-1} \binom{n-1}{j} p^j (1-p)^{n-1-j} = 1
$

### Thus,

$
E[Y] = n p \cdot 1 = np
$

Therefore, the expectation of the binomial random variable $Y$ is:

$
E[Y] = np
$

*Answer:*

The general formula for the expected value of this distribution, which is a Poisson distribution, is:
Find a general formula for the expected value of this distribution.

\begin{aligned}
E[Y] &= \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda} \\
&= \sum_{y=1}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda} \quad \text{(since $y = 0$ term is 0)} \\
&= \sum_{y=1}^{\infty} \frac{\lambda^y}{(y-1)!} e^{-\lambda} \\
&= e^{-\lambda} \sum_{y=1}^{\infty} \frac{\lambda^y}{(y-1)!} \quad \text{(pulling out $e^{-\lambda}$ as a constant)} \\
&= e^{-\lambda} \sum_{y=0}^{\infty} \frac{\lambda^{y+1}}{y!} \quad \text{(reindexing the sum)} \\
&= e^{-\lambda} \cdot \lambda \sum_{y=0}^{\infty} \frac{\lambda^y}{y!} \quad \text{(factor out $\lambda$ from the sum)} \\
&= e^{-\lambda} \cdot \lambda \cdot e^{\lambda} \quad \text{(using the Taylor series expansion of $e^x$)} \\
&= \lambda \quad \text{(simplify)}
\end{aligned}


**Question 2: A certain type of tree has seedlings randomly dispersed in a large area, with the mean density of seedlings being approximately five per square yard. If a forester randomly locates ten 1-square-yard sampling regions in the area, find the probability that none of the regions will contain seedlings.**

*Answer:*

The probability of a region not containing any seedlings is the complement of the probability of it containing at least one seedling. Let $p$ be the probability that a region contains at least one seedling. Then, the probability that none of the regions contain seedlings is $1 - p$.

Given that the mean density of seedlings is approximately five per square yard, the probability of a square yard containing no seedlings is $e^{-5}$ (since this follows a Poisson distribution with mean $\lambda = 5$).

Therefore, the probability that a region of one square yard contains at least one seedling is $1 - e^{-5}$.

Since the sampling regions are independent, the probability that none of the ten regions contain seedlings is $(1 - e^{-5})^{10}$.

Thus, the probability that none of the regions will contain seedlings is:

$$ 1 - (1 - e^{-5})^{10} $$

This expression evaluates to approximately 0.00674.
